## __Exercice 1__ : __Le Laboratoire de Crash Test__

- Crée une fonction nommée `diviser_deux_nombres()`.
- À l'intérieur, demande à l'utilisateur de saisir un premier nombre (le numérateur) puis un deuxième nombre (le dénominateur).
- Le programme doit afficher le résultat de la division.
- Le __Défi__ : Ton code doit être totalement indestructible. Tu dois anticiper et capturer deux erreurs différentes dans le même bloc :
  * Si l'utilisateur tape du texte au lieu d'un nombre (Intercepter _`ValueError`_).
  * Si l'utilisateur tape le chiffre 0 comme deuxième nombre (Intercepter _`ZeroDivisionError`_).
- Ajoute un bloc finally qui affiche "Fin de l'opération de calcul." à chaque tentative.


## __Exercice 2__ : __La Douane Algorithmique__
- Étape 1 (Le Prix) : Demande à l'utilisateur de saisir le prix d'un article en USD. La saisie doit obligatoirement être un nombre à virgule (float). Tant que l'utilisateur tape du texte ou un prix inférieur ou égal à 0, le programme doit lui afficher une erreur et lui redemander le prix.
- Étape 2 (La Quantité) : Une fois le prix validé, le programme passe à l'étape suivante et demande la quantité d'articles achetés. La saisie doit obligatoirement être un entier (int) supérieur à 0. Tant que ce n'est pas correct, le programme boucle.
- Étape 3 (Le Résultat) : Une fois les deux barrières franchies, le programme affiche enfin le montant total à payer.

## __Exercice 3__ : __Le Guichet Automatique Sécurisé__

__Le contexte__ : Tu codes le système de retrait d'argent d'une banque. Le client possède un solde initial fixe de 500 USD sur son compte. Ton programme doit lui demander combien il souhaite retirer, valider sa saisie, vérifier les règles métiers, et mettre à jour son solde.
1. La __Boucle de Saisie__ : Le programme demande à l'utilisateur : "**_Combien voulez-vous retirer (USD) ?_**" tant que la saisie n'est pas valide.
2. La __Sécurité Technique__ (_`try/except/as e`_) : Si l'utilisateur tape du texte ou des symboles (ex: "cinquante"), le programme doit capturer la `ValueError` grâce à un alias `as e`, afficher un message d'erreur clair contenant l'explication, et lui redemander le montant.
3. La __Sécurité Métier__ (raise) : Si l'utilisateur saisit un nombre valide, tu dois appliquer deux règles bancaires :
   - __Règle A__ : Le montant du retrait ne peut pas être négatif ou égal à zéro. Si c'est le cas, tu dois provoquer volontairement une `ValueError` avec le message : "**_Le montant doit être supérieur à 0 USD._**"
   - __Règle B__ : Le montant ne peut pas dépasser le solde disponible (500 USD). Si c'est le cas, tu dois provoquer volontairement une `ValueError` avec le message : "**_Solde insuffisant pour effectuer ce retrait._**"
4. La __Validation__ (_`break`_) : Si le montant passe toutes les étapes sans lever d'erreur, le solde est mis à jour, on affiche le nouveau solde et on sort de la boucle.
