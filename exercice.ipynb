## __Exercice 3__ : __Le Guichet Automatique Sécurisé__

__Le contexte__ : Tu codes le système de retrait d'argent d'une banque. Le client possède un solde initial fixe de 500 USD sur son compte. Ton programme doit lui demander combien il souhaite retirer, valider sa saisie, vérifier les règles métiers, et mettre à jour son solde.
1. La __Boucle de Saisie__ : Le programme demande à l'utilisateur : "**_Combien voulez-vous retirer (USD) ?_**" tant que la saisie n'est pas valide.
2. La __Sécurité Technique__ (_`try/except/as e`_) : Si l'utilisateur tape du texte ou des symboles (ex: "cinquante"), le programme doit capturer la `ValueError` grâce à un alias `as e`, afficher un message d'erreur clair contenant l'explication, et lui redemander le montant.
3. La __Sécurité Métier__ (raise) : Si l'utilisateur saisit un nombre valide, tu dois appliquer deux règles bancaires :
   - __Règle A__ : Le montant du retrait ne peut pas être négatif ou égal à zéro. Si c'est le cas, tu dois provoquer volontairement une `ValueError` avec le message : "**_Le montant doit être supérieur à 0 USD._**"
   - __Règle B__ : Le montant ne peut pas dépasser le solde disponible (500 USD). Si c'est le cas, tu dois provoquer volontairement une `ValueError` avec le message : "**_Solde insuffisant pour effectuer ce retrait._**"
4. La __Validation__ (_`break`_) : Si le montant passe toutes les étapes sans lever d'erreur, le solde est mis à jour, on affiche le nouveau solde et on sort de la boucle.

### __Illustration__

```console
=== BIENVENUE À LA BANQUE ===
Votre solde actuel est de : 500 USD

Combien voulez-vous retirer (USD) ?
-> cent
❌ Erreur Saisie : invalid literal for int() with base 10: 'cent'
Veuillez entrer un nombre entier.

Combien voulez-vous retirer (USD) ?
-> -50
❌ Erreur Métier : Le montant doit être supérieur à 0 USD.

Combien voulez-vous retirer (USD) ?
-> 600
❌ Erreur Métier : Solde insuffisant pour effectuer ce retrait.

Combien voulez-vous retirer (USD) ?
-> 200
🎉 Retrait réussi ! Vous avez retiré 200 USD.
Votre nouveau solde est de : 300 USD.
```

## __Exercice 4 : Le Panier d'Achat Électronique__

**Le contexte :** Tu codes le système de panier d'achat pour un site de commerce en ligne. Chaque article ajouté au panier possède un nom et un prix. Le panier doit être capable de calculer le total tout seul.

### 📋 __Le Cahier des Charges__ :

Créer un script `panier.py` contenant une seule classe :

1. **La Classe `Article` :**
* **Le Constructeur (`__init__`) :** Il doit recevoir deux informations : le `nom` de l'article (ex: "Clavier") et son `prix` (ex: 25).
* **Les Attributs :** Stocker ces informations dans `self.nom` et `self.prix`.


2. **La Classe `Panier` :**
* **Le Constructeur (`__init__`) :** Il ne reçoit aucun paramètre à la naissance (à part `self`). À l'intérieur, il doit initialiser un attribut `self.liste_articles` égal à une **liste vide `[]`**.
* **Méthode 1** : `ajouter_article(self, nouvel_article)` : Cette méthode prend un objet `Article` en paramètre et l'ajoute (avec `.append()`) à sa liste. Elle doit afficher un message de confirmation.
* **Méthode 2** : `afficher_contenu(self)` : Elle fait une boucle pour afficher le nom et le prix de chaque article présent dans le panier.
* **Méthode 3** : `calculer_total(self)` : Elle parcourt les articles du panier, additionne leurs prix, et affiche le montant total à payer.


### __Illustration__ 

#### Code source
```python
# 1. Création des articles (Objets de la classe Article)
article1 = Article("Souris Gaming", 45)
article2 = Article("Tapis de souris", 15)
article3 = Article("Casque Audio", 80)

# 2. Création du panier (Objet de la classe Panier)
mon_panier = Panier()

# 3. Utilisation des méthodes du panier
mon_panier.ajouter_article(article1)
mon_panier.ajouter_article(article2)
mon_panier.ajouter_article(article3)

print("\n--- CONTENU DE MON PANIER ---")
mon_panier.afficher_contenu()

print("\n--- CAISSE ---")
mon_panier.calculer_total()

```


## **Exercice 5 : Le Gestionnaire de Téléchargements**

**Le contexte :** Tu codes un logiciel qui gère des téléchargements de fichiers (comme une mise à jour d'application ou une vidéo). Chaque fichier a une taille, un nom, et un statut ("En attente", "En cours", "Terminé").

### 📋 __Le Cahier des Charges__ :

Créer un script `telechargement.py` contenant une seule classe, la classe `Fichier` :
* **Le Constructeur (`__init__`) :** Il reçoit deux informations : le `nom` (ex: `"ubuntu.iso"`) et la `taille_totale` en Mo (ex: `1000`).
  
* **Les Attributs à initialiser :**
  - `self.nom` = le nom reçu.
  - `self.taille_totale` = la taille reçue.
  - `self.taille_telechargee` = **`0`** (au début, on n'a encore rien téléchargé).
  - `self.statut` = **`"En attente"`** (le statut de départ par défaut).


* **Méthode 1 : `afficher_progression(self)`** :
Cette méthode doit afficher le nom du fichier, son statut, et combien de Mo ont été téléchargés par rapport au total.
*(Ex d'affichage : `[En cours] ubuntu.iso -> 300 / 1000 Mo`)*

* **Méthode 2 : `telecharger(self, mo_recus)`** :
   - Elle doit ajouter les `mo_recus` à l'attribut `self.taille_telechargee`.
   - Elle doit passer le statut à `"En cours"`.
   - **La Condition :** Si `self.taille_telechargee` devient supérieure ou égale à `self.taille_totale`, alors le téléchargement est fini ! On bloque la valeur au maximum de la taille totale, et on passe le statut à `"Terminé"`.
   - À la toute fin, elle doit appeler elle-même la méthode `self.afficher_progression()` pour montrer l'évolution à l'utilisateur.


### __Illustration__

#### Code source :

```python
# 1. Création du fichier (1000 Mo à télécharger)
video_cours = Fichier("tuto_poo.mp4", 1000)

# 2. On vérifie l'état initial
print("--- ÉTAT INITIAL ---")
video_cours.afficher_progression()

print("\n--- DEBUT DU TELECHARGEMENT ---")
# 3. Le téléchargement commence (on reçoit un premier bloc de 400 Mo)
video_cours.telecharger(400)

# 4. On reçoit un deuxième bloc de 400 Mo
video_cours.telecharger(400)

# 5. On reçoit le dernier bloc de 300 Mo (ce qui dépasse le total)
video_cours.telecharger(300)

```

#### Le rendu attendu dans le terminal :

```text
--- ÉTAT INITIAL ---
[En attente] tuto_poo.mp4 -> 0 / 1000 Mo

--- DEBUT DU TELECHARGEMENT ---
📥 Progression : [En cours] tuto_poo.mp4 -> 400 / 1000 Mo
📥 Progression : [En cours] tuto_poo.mp4 -> 800 / 1000 Mo
📥 Progression : [Terminé] tuto_poo.mp4 -> 1000 / 1000 Mo 🎉 Téléchargement complété !

```
